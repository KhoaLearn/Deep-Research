```python
# Convert ticker symbols to lowercase
vn30_tickers_lower = [ticker.lower() for ticker in VN30_TICKERS]

# Iterate over the list of lowercase tickers and fetch data
all_stock_data = {}
for ticker in vn30_tickers_lower:
    print(f"Processing ticker: {ticker}")
    company = Company(ticker)
    all_stock_data[ticker] = company.stock_price
```

```python
# Join all DataFrames on the Date column using left join
combined_df = None
for ticker, df in all_stock_data.items():
    if combined_df is None:
        combined_df = df
    else:
        combined_df = combined_df.join(df, how='left')
```
